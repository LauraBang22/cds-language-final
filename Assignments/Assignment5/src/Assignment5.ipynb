### Which assignment generated the most emissions in terms of CO₂eq? Explain why this might be?

I start by loading the dataframe that contains the CO₂eq per script. 

```python
filename = os.path.join("..", "emissions", "examples", "emissions.csv")
data = pd.read_csv(filename)
```

```python
#defining colorpalettes to use in the plots
colors1 = ["#C9E4DE", "#F7D9C4", "#F7D9C4", "#F7D9C4", "#F2C6DE", "#C6DEF1", "#C6DEF1"]
colors2 = ["#C9E4DE", "#F7D9C4", "#F2C6DE", "#C6DEF1"]
```

As predicted, this plot shows that the assignment with the highest emission is assignment 4. # How robust do you think these results are and how/where might they be improved?

To check how robust the results are, I have tried running assignment 1 multiple times. I have run the code five seperate times, and have then plotted the total emissions for each time.

```python
filename = os.path.join("..", "emissions", "examples", "emissions1.csv")
assignment1 = pd.read_csv(filename)

plt.figure(figsize=(8, 6))  
assignment1.plot(x="project_name", y="emissions", kind = "bar", color = "#DBCDF0", legend=None)


plt.xlabel("scripts")
plt.ylabel("emissions")
plt.title('Distribution of values in emissions across assignment 1 - multiple times')

plt.xticks(rotation=65)  
plt.tight_layout()
plt.savefig("../out/robustness.png")
plt.show()
```

As you can see, there are slight variations in how much CO₂eq it emits, for each run. I wondered why that was, because the code does the same exact thing every time. I therefore decided to also plot the duration of the runs to see if, that might have had an impact. The duration is measured in seconds.

```python
plt.figure(figsize=(10, 6)) 

grid = plt.GridSpec(1, 2, wspace=0.3, hspace=0.3)  

ax1 = plt.subplot(grid[0])
assignment1.plot(kind="bar", x="project_name", y="emissions", color="#DBCDF0", legend=None, ax=ax1)
ax1.set_xlabel("Project Name")
ax1.set_ylabel("Emissions")
ax1.set_title("Emissions")  

ax2 = plt.subplot(grid[1])
assignment1.plot(kind="bar", x="project_name", y="duration", color="#DBCDF0", legend=None, ax=ax2)
ax2.set_xlabel("Project Name")
ax2.set_ylabel("Duration")
ax2.set_title("Duration") 

plt.tight_layout()
plt.savefig("../out/duration.png")
plt.show()
```

As you can see, the plots for emission and duration are very similar. Which indicates that how long it takes to run the code has an impact on how much emission it generates. Since the duration of the run can vary, based on what else the computer is doing and in this case how strained Ucloud is, it makes sense that the amount of emission also varies a little bit. It is also very little that it varies, so I would say that the way it measures it, is pretty robust.

All of these runs are done on a small machine with 4 CPU. Another thing that I imagine might factor into the emission is how large a machine that the code is run on.

When it comes to ways you can limit your emission when running code, and therefore improve your emission results, there is a few things to be done. Especially when writing  and testing the code, is probably where you can do the most. You can consider exactly want you want the code to do, before you start writing it, instead of just trying a bunch of different things. You can also limit the amount of data, you run the code on to test if it works. If you only want to see if the code you have written, can acutally run, then there is no reason to test it on the entire dataset, just a small slice of it will suffice. Also be mindful of how many times, you feel it is necessary to test your code.
